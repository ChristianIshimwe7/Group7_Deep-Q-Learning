````python
"""
train.py
========
DQN agent training script for ALE/Breakout-v5 using Stable Baselines 3.

Usage
-----
# Run all 10 experiments for a specific member (1-4)
    python train.py --member 1

# Run a single experiment (e.g. experiment 3 for member 2)
    python train.py --member 2 --exp 3

# Quick smoke-test with fewer timesteps
    python train.py --member 1 --exp 1 --timesteps 50000
"""

import os
import json
import argparse
import numpy as np
import gymnasium as gym
import ale_py

from stable_baselines3 import DQN
from stable_baselines3.common.atari_wrappers import AtariWrapper
from stable_baselines3.common.vec_env import DummyVecEnv, VecFrameStack
from stable_baselines3.common.callbacks import EvalCallback, BaseCallback

gym.register_envs(ale_py)

# ---------------------------------------------------------------------------
# Callback – logs mean reward every N steps
# ---------------------------------------------------------------------------
class TrainingLogger(BaseCallback):
    def __init__(self, log_interval: int = 10_000):
        super().__init__()
        self.log_interval = log_interval
        self.ep_rewards: list[float] = []
        self.ep_lengths: list[int]   = []
        self._cur_rew: dict[int, float] = {}
        self._cur_len: dict[int, int]   = {}

    def _on_step(self) -> bool:
        for i, (done, rew) in enumerate(
            zip(self.locals["dones"], self.locals["rewards"])
        ):
            self._cur_rew[i] = self._cur_rew.get(i, 0.0) + float(rew)
            self._cur_len[i] = self._cur_len.get(i, 0)   + 1
            if done:
                self.ep_rewards.append(self._cur_rew[i])
                self.ep_lengths.append(self._cur_len[i])
                self._cur_rew[i] = 0.0
                self._cur_len[i] = 0

        if self.n_calls % self.log_interval == 0 and self.ep_rewards:
            last = self.ep_rewards[-20:]
            print(
                f"  [step {self.num_timesteps:>8,}] "
                f"mean_reward(last 20 eps)={np.mean(last):.2f}  "
                f"episodes={len(self.ep_rewards)}"
            )
        return True


# ---------------------------------------------------------------------------
# Environment helpers
# ---------------------------------------------------------------------------
def make_env_fn(render_mode=None):
    def _init():
        env = gym.make("ALE/Breakout-v5", render_mode=render_mode)
        env = AtariWrapper(env)
        return env
    return _init

def build_env(render_mode=None, n_stack=4):
    return VecFrameStack(DummyVecEnv([make_env_fn(render_mode)]), n_stack=n_stack)


# ---------------------------------------------------------------------------
# Core training function
# ---------------------------------------------------------------------------
def run_experiment(
    *,
    member:                  int,
    exp_num:                 int,
    experiment_name:         str,
    policy:                  str   = "CnnPolicy",
    learning_rate:           float = 1e-4,
    gamma:                   float = 0.99,
    batch_size:              int   = 32,
    exploration_initial_eps: float = 1.0,
    exploration_final_eps:   float = 0.01,
    exploration_fraction:    float = 0.10,
    total_timesteps:         int   = 500_000,
):
    tag        = f"M{member}_E{exp_num:02d}_{experiment_name}"
    model_path = f"models/{tag}"
    best_path  = f"best/{tag}"
    log_path   = f"logs/{tag}"
    tb_path    = f"tensorboard/{tag}"

    for d in [model_path, best_path, log_path, tb_path]:
        os.makedirs(d, exist_ok=True)

    print(f"\n{'='*65}")
    print(f"  Member {member}  |  Experiment {exp_num:02d}  |  {experiment_name}")
    print(f"  policy={policy}  lr={learning_rate}  gamma={gamma}  "
          f"batch={batch_size}")
    print(f"  eps: {exploration_initial_eps} → {exploration_final_eps} "
          f"over {exploration_fraction*100:.0f}% of training")
    print(f"{'='*65}\n")

    train_env = build_env()
    eval_env  = build_env()

    model = DQN(
        policy                  = policy,
        env                     = train_env,
        learning_rate           = learning_rate,
        buffer_size             = 100_000,
        learning_starts         = 10_000,
        batch_size              = batch_size,
        gamma                   = gamma,
        train_freq              = 4,
        target_update_interval  = 1_000,
        exploration_fraction    = exploration_fraction,
        exploration_initial_eps = exploration_initial_eps,
        exploration_final_eps   = exploration_final_eps,
        optimize_memory_usage   = False,
        verbose                 = 0,
        tensorboard_log         = tb_path,
    )

    logger    = TrainingLogger(log_interval=10_000)
    eval_cb   = EvalCallback(
        eval_env,
        best_model_save_path = best_path,
        log_path             = log_path,
        eval_freq            = 25_000,
        n_eval_episodes      = 5,
        deterministic        = True,
        render               = False,
    )

    model.learn(
        total_timesteps = total_timesteps,
        callback        = [logger, eval_cb],
        progress_bar    = True,
    )

    model.save(f"{model_path}/dqn_model")
    model.save("dqn_model")

    mean_rew = float(np.mean(logger.ep_rewards[-20:])) if logger.ep_rewards else 0.0
    meta = dict(
        tag=tag, policy=policy,
        learning_rate=learning_rate, gamma=gamma, batch_size=batch_size,
        exploration_initial_eps=exploration_initial_eps,
        exploration_final_eps=exploration_final_eps,
        exploration_fraction=exploration_fraction,
        total_timesteps=total_timesteps,
        mean_reward_last20=mean_rew,
        total_episodes=len(logger.ep_rewards),
    )
    with open(f"{log_path}/meta.json", "w") as f:
        json.dump(meta, f, indent=2)

    print(f"\n  ✓ Model saved → {model_path}/dqn_model.zip")
    print(f"  ✓ Mean reward (last 20 eps): {mean_rew:.2f}\n")

    train_env.close()
    eval_env.close()
    return mean_rew


# ---------------------------------------------------------------------------
# 40 experiments  (10 per member)
# ---------------------------------------------------------------------------
ALL_EXPERIMENTS = {

    # ── MEMBER 1 ────────────────────────────────────────────────────────────
    1: [
        dict(experiment_name="baseline",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="high_lr",
             policy="CnnPolicy", learning_rate=1e-3, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="low_lr",
             policy="CnnPolicy", learning_rate=5e-5, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="gamma_95",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.95,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="gamma_999",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.999,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="batch_64",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=64, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="batch_16",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=16, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="eps_end_05",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.05, exploration_fraction=0.10),
        dict(experiment_name="slow_decay",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.20),
        dict(experiment_name="mlp_policy",
             policy="MlpPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
    ],

    # ── MEMBER 2 ────────────────────────────────────────────────────────────
    2: [
        dict(experiment_name="lr_2e4",
             policy="CnnPolicy", learning_rate=2e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="gamma_98",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.98,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="batch_128",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=128, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="eps_start_05",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=0.5,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="eps_end_001",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.001, exploration_fraction=0.10),
        dict(experiment_name="lr_5e4",
             policy="CnnPolicy", learning_rate=5e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="combined_A",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.97,
             batch_size=64, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.15),
        dict(experiment_name="combined_B",
             policy="CnnPolicy", learning_rate=2e-4, gamma=0.98,
             batch_size=64, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="fast_decay",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.02, exploration_fraction=0.05),
        dict(experiment_name="lr_3e4",
             policy="CnnPolicy", learning_rate=3e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
    ],

    # ── MEMBER 3 ────────────────────────────────────────────────────────────
    3: [
        dict(experiment_name="baseline_m3",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="gamma_90",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.90,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="batch_256",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=256, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="eps_end_10",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.10, exploration_fraction=0.10),
        dict(experiment_name="frac_30",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.30),
        dict(experiment_name="lr_75e5",
             policy="CnnPolicy", learning_rate=7.5e-5, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="gamma_995",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.995,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="batch_48",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=48, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="best_combined_m3",
             policy="CnnPolicy", learning_rate=2e-4, gamma=0.995,
             batch_size=64, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.12),
        dict(experiment_name="mlp_tuned",
             policy="MlpPolicy", learning_rate=5e-4, gamma=0.99,
             batch_size=64, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
    ],

    # ── MEMBER 4 ────────────────────────────────────────────────────────────
    4: [
        dict(experiment_name="baseline_m4",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="eps_start_08",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=0.8,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="frac_08",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.08),
        dict(experiment_name="lr_15e4",
             policy="CnnPolicy", learning_rate=1.5e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="gamma_985",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.985,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="eps_end_005",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.005, exploration_fraction=0.10),
        dict(experiment_name="batch_96",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=96, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="lr_8e5",
             policy="CnnPolicy", learning_rate=8e-5, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.10),
        dict(experiment_name="frac_25",
             policy="CnnPolicy", learning_rate=1e-4, gamma=0.99,
             batch_size=32, exploration_initial_eps=1.0,
             exploration_final_eps=0.01, exploration_fraction=0.25),
        dict(experiment_name="best_combined_m4",
             policy="CnnPolicy", learning_rate=1.5e-4, gamma=0.995,
             batch_size=48, exploration_initial_eps=1.0,
             exploration_final_eps=0.005, exploration_fraction=0.12),
    ],
}


# ---------------------------------------------------------------------------
# Entry point
# ---------------------------------------------------------------------------
if __name__ == "__main__":
    parser = argparse.ArgumentParser()
    parser.add_argument("--member",    type=int, required=True, choices=[1,2,3,4])
    parser.add_argument("--exp",       type=int, default=None)
    parser.add_argument("--timesteps", type=int, default=500_000)
    args = parser.parse_args()

    exps = ALL_EXPERIMENTS[args.member]
    if args.exp is not None:
        exps      = [exps[args.exp - 1]]
        start_num = args.exp
    else:
        start_num = 1

    results = {}
    for i, cfg in enumerate(exps, start=start_num):
        reward = run_experiment(member=args.member, exp_num=i,
                                total_timesteps=args.timesteps, **cfg)
        results[f"E{i:02d}_{cfg['experiment_name']}"] = reward

    print("\n" + "="*65)
    print(f"  SUMMARY  –  Member {args.member}")
    print("="*65)
    for name, r in results.items():
        print(f"  {name:<40}  mean_reward={r:.2f}")
    print("="*65)
````

# this our play.py

````python
"""
play.py
=======
Load the best trained DQN model and play ALE/Breakout-v5.

Uses a Greedy Q-Policy (deterministic=True) so the agent always
picks the action with the highest Q-value.

Usage
-----
    python play.py                          # GUI, 5 episodes, load dqn_model.zip
    python play.py --model best/my_run/best_model
    python play.py --episodes 10
    python play.py --no-render --episodes 20   # headless / server
"""

import argparse
import time
import numpy as np
import gymnasium as gym
import ale_py

from stable_baselines3 import DQN
from stable_baselines3.common.atari_wrappers import AtariWrapper
from stable_baselines3.common.vec_env import DummyVecEnv, VecFrameStack

gym.register_envs(ale_py)


def get_args():
    p = argparse.ArgumentParser()
    p.add_argument("--model",     type=str,   default="dqn_model")
    p.add_argument("--episodes",  type=int,   default=5)
    p.add_argument("--no-render", action="store_true")
    p.add_argument("--delay",     type=float, default=0.02)
    return p.parse_args()


def make_env_fn(render_mode=None):
    def _init():
        env = gym.make("ALE/Breakout-v5", render_mode=render_mode)
        env = AtariWrapper(env)
        return env
    return _init

def build_env(render_mode=None, n_stack=4):
    return VecFrameStack(DummyVecEnv([make_env_fn(render_mode)]), n_stack=n_stack)


def load_model(path: str, env):
    candidates = [
        path,
        f"best/{path}/best_model",
        f"models/{path}/dqn_model",
        "dqn_model",
    ]
    for c in candidates:
        try:
            m = DQN.load(c, env=env)
            print(f"  ✓ Model loaded from: {c}.zip")
            return m
        except Exception:
            pass
    raise FileNotFoundError("No model found. Run train.py first.")


def play(model_path, num_episodes, render, delay):
    render_mode = "human" if render else None
    env         = build_env(render_mode=render_mode)
    model       = load_model(model_path, env)

    print(f"\n{'─'*60}")
    print(f"  Environment : ALE/Breakout-v5")
    print(f"  Policy      : {type(model.policy).__name__}  (Greedy Q)")
    print(f"  Episodes    : {num_episodes}")
    print(f"{'─'*60}\n")

    ep_rewards, ep_lengths = [], []

    for ep in range(1, num_episodes + 1):
        obs       = env.reset()
        done      = False
        total_rew = 0.0
        steps     = 0

        while not done:
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, done, info = env.step(action)
            total_rew += float(reward[0])
            steps     += 1
            if render:
                time.sleep(delay)
            if done[0]:
                break

        ep_rewards.append(total_rew)
        ep_lengths.append(steps)
        print(f"  Episode {ep:>3}/{num_episodes} │ "
              f"steps={steps:>6,} │ reward={total_rew:>7.2f}")

    env.close()

    print(f"\n{'═'*60}")
    print(f"  Mean reward : {np.mean(ep_rewards):>8.2f}")
    print(f"  Std  reward : {np.std(ep_rewards):>8.2f}")
    print(f"  Max  reward : {np.max(ep_rewards):>8.2f}")
    print(f"  Min  reward : {np.min(ep_rewards):>8.2f}")
    print(f"  Mean length : {np.mean(ep_lengths):>8.1f} steps")
    print(f"{'═'*60}\n")


if __name__ == "__main__":
    args = get_args()
    play(args.model, args.episodes, not args.no_render, args.delay)
```

---

## 📄 `requirements.txt`
```
gymnasium[atari,accept-rom-license]
stable-baselines3[extra]
ale-py
torch
tensorboard
numpy
````
